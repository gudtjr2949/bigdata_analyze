```python
# install chromium, its driver, and selenium
!apt-get update
!apt install chromium-chromedriver
!cp /usr/lib/chromium-browser/chromedriver /usr/bin
!pip install selenium
# set options to be headless, ..
from selenium import webdriver
options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
# open it, go to a website, and get results
driver = webdriver.Chrome('chromedriver',options=options)
driver.get("https://search.naver.com/search.naver?where=news&sm=tab_jum&query=%EB%84%B7%EB%A7%88%EB%B8%94") # 원하는 기업 검색하고 뉴스 창 드간 다음 주소 복붙
print(driver.page_source)  # results
```

```python
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
from tqdm import tqdm
import time

# 옵션 클릭

elem_option = driver.find_element_by_xpath('//*[@id="snb"]/div[1]/div/div[2]/a')
elem_option.send_keys('\n')

# 직접입력 클릭
elem_input = driver.find_element_by_xpath('//*[@id="snb"]/div[2]/ul/li[2]/div/div[1]/a[9]')
elem_input.send_keys('\n')

# 4월 클릭
elem_month = driver.find_element_by_xpath('//*[@id="snb"]/div[2]/ul/li[2]/div/div[3]/div[2]/div[2]/div/div/div/ul/li[4]/a')
elem_month.send_keys('\n')

# 날짜 클릭
elem_day = driver.find_element_by_xpath('//*[@id="snb"]/div[2]/ul/li[2]/div/div[3]/div[2]/div[3]/div/div/div/ul/li[1]/a')
elem_day.send_keys('\n')

# 적용 버튼 클릭
elem_apply = driver.find_element_by_xpath('//*[@id="snb"]/div[2]/ul/li[2]/div/div[3]/div[3]/button')
elem_apply.send_keys('\n')

html_source = driver.page_source
soup = BeautifulSoup(html_source, 'html.parser')

title_1 = soup.select("a.news_tit")
title_final_1 = []

# 4월 1일 ~ 4월 6일 까지 기사 제목 크롤링
for i in range(len(title_1)):
    temp_title = title_1[i].text
    temp_title = temp_title.replace('\n', '')
    temp_title = temp_title.replace('\t', '')
    temp_title = temp_title.replace('    ', '')
    title_final_1.append(temp_title)
    

# 날짜 변경 버튼 클릭
elem_change = driver.find_element_by_xpath('//*[@id="snb"]/div[2]/ul/li[2]/div/div[1]/a[9]')
elem_change.send_keys('\n')


# 시작 시간 변경
elem_time = driver.find_element_by_xpath('//*[@id="snb"]/div[2]/ul/li[2]/div/div[3]/div[1]/span[3]/a')
elem_time.send_keys('\n')

# 3월
elem_3 = driver.find_element_by_xpath('//*[@id="snb"]/div[2]/ul/li[2]/div/div[3]/div[2]/div[2]/div/div/div/ul/li[3]/a')
elem_3.send_keys('\n')

# 24일
elem_d = driver.find_element_by_xpath('//*[@id="snb"]/div[2]/ul/li[2]/div/div[3]/div[2]/div[3]/div/div/div/ul/li[24]/a')
elem_d.send_keys('\n')

elem_apply = driver.find_element_by_xpath('//*[@id="snb"]/div[2]/ul/li[2]/div/div[3]/div[3]/button')
elem_apply.send_keys('\n')

html_source = driver.page_source
soup = BeautifulSoup(html_source, 'html.parser')

title_2 = soup.select("a.news_tit")
title_final_2 = []

# 3월 24일 ~ 4월 1일 까지 기사 제목 크롤링
for i in range(len(title_2)):
    temp_title = title_2[i].text
    temp_title = temp_title.replace('\n', '')
    temp_title = temp_title.replace('\t', '')
    temp_title = temp_title.replace('    ', '')
    title_final_2.append(temp_title)
```

```python
from konlpy.tag import Twitter
```

```python
text = str(title_final_1) # 일단 4월1일부터 6일까지만 분석함
```

```python
df = []

def keyword_extractor(tagger, text):
    tokens = tagger.phrases(text)
    tokens = [ token for token in tokens if len(token) > 1 ] # 한 글자인 단어는 제외
    count_dict = [(token, text.count(token)) for token in tokens ]
    ranked_words = sorted(count_dict, key=lambda x:x[1], reverse=True)[:10]
    return [ keyword for keyword, freq in ranked_words ]

if __name__ == '__main__':
    twit = Twitter()
    print( keyword_extractor(twit, text) )
    df = keyword_extractor(twit, text)
```
