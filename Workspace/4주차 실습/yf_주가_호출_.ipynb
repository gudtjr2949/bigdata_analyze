```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
GSretail_df = yf.download('007070.KS',
                      start='2022-03-01',
                      end='2022-04-05',
                      progress=False)
```

```python
GSretail_df 
```

```python
GSretail_df  = GSretail_df [["Close"]]

GSretail_df  = GSretail_df .reset_index()

GSretail_df.columns = ['day', 'price']

GSretail_df['day'] = pd.to_datetime(GSretail_df['day'])

GSretail_df.index = GSretail_df['day']
GSretail_df.set_index('day', inplace=True)

GSretail_df
```
